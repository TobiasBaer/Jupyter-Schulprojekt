# Ein erster Anwendungsfall - Kreisberechnung

Mit dem folgenden Programm kann man beispielsweise die Fläche eines Kreises berechnen:


```python
def area_of_circle( radius ):
  pi = 3.14
  return pi * radius * radius
```

Führt man diese Funktion aus, kann man folgende Ausgabe erzeugen:

```python
area = area_of_circle( 10 )
print(f"Die Fläche eines Kreises mit einem Radius von 10 cm beträgt {area} cm2")
```

Jupyter Notebooks sind dabei interaktiv. Das heißt man kann z.B. Werte durch Schieberegler ändern und sich Schaubilder live generieren lassen.

```python
import matplotlib.pyplot as plt
x = [ radius for radius in range(10) ]
y = [ area_of_circle( radius ) for radius in range(10) ]
plt.plot(x, y)
plt.xlabel("Radius")
plt.ylabel("Flächeninhalt")
plt.title("Beziehung Radius zu Flächeninhalt")
plt.show()
```
